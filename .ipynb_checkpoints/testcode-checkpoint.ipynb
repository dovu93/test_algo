```python
import pandas as pd
from pathlib import Path
import seaborn as sns
import numpy as np
import matplotlib
```

```python
path = Path("Resources/portfolio.csv")
path1 = Path("Resources/idex_ticker.csv")
df = pd.read_csv(path, index_col = "Date", infer_datetime_format = True, parse_dates = True)
df1 = pd.read_csv(path1, index_col = "Date", infer_datetime_format = True, parse_dates = True)
```

```python
df1
```

```python
df.drop(columns = ["Date.1", "Date.2", "Date.3"], inplace = True)
df.head()
```

```python
amzn = df["AMZN"]
bp = df["BP"]
dis = df["DIS"]
nymt = df["NYMT"]
```

```python
a = amzn.plot(figsize = (25,20))
amzn.rolling(window = 50).mean().plot(ax = a)
amzn.rolling(window = 200).mean().plot(ax = a)

a.legend(["AMZN", "50 Day Average", "200 Day Average"])
```

```python
b = dis.plot(figsize = (25,20))
dis.rolling(window = 10).mean().plot(ax = b)
dis.rolling(window = 50).mean().plot(ax = b)
dis.rolling(window = 100).mean().plot(ax = b)
dis.rolling(window = 200).mean().plot(ax = b)

b.legend(["DIS", "10 Day Average", "50 Day Average", "100 Day Average", "200 Day Average"])
```

```python
d = df1.plot(figsize = (25,20))
df1.rolling(window = 10).mean().plot(ax = d)
df1.rolling(window = 50).mean().plot(ax = d)
df1.rolling(window = 100).mean().plot(ax = d)
df1.rolling(window = 200).mean().plot(ax = d)

d.legend(["IDEX", "10 Day Average", "50 Day Average", "100 Day Average", "200 Day Average"])
```
